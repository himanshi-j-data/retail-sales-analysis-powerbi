Remove Duplicates

```python
df_cleaned = df.drop_duplicates()
```

Handle Missing Values

```python
df_cleaned = df_cleaned.dropna()
```

Confirm after Cleaning

```python
df_cleaned.shape
df_cleaned.isnull().sum()
df_cleaned.duplicated().sum()
```

Save Cleaned File

```python
df_cleaned.to_csv("Superstore_Cleaned.csv", index=False)
```

```python
df_cleaned.isnull().sum()
```

```python
# As sum of null values is Zero that means data is cleaned
```

```python
# Convert date columns
```

```python
df_cleaned["Order Date"] = pd.to_datetime(df_cleaned["Order Date"])
df_cleaned["Ship Date"] = pd.to_datetime(df_cleaned["Ship Date"])
```

```python
df.dtypes
```

```python
# Create New Columns (Feature Engineering)
```

```python
df_cleaned["Profit Margin"] = df_cleaned["Profit"]/df_cleaned["Sales"]
```

```python
df_cleaned["Order Year"] = df_cleaned["Order Date"].dt.year
df_cleaned["Order Month Name"] = df_cleaned["Order Date"].dt.month_name()
```

```python
df_cleaned["Order Month"] = df_cleaned["Order Date"].dt.month
```

```python
df_cleaned
```

```python
# Top 5 Customer by Revenue
```

```python
top_5_customers = (
    df_cleaned.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Customers by Revenue:")
print(top_5_customers)
```

```python
# Most Profitable Category
```

```python
category_profit = (
    df_cleaned.groupby("Category")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

print("Profit by Category:")
print(category_profit)
```

```python
# Region with Highest sales
```

```python
region_sales = (
    df_cleaned.groupby("Region")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

print("Sales by Region:")
print(region_sales)
```

```python
# Monthly Revenue Trend
```

```python
monthly_revenue = (
    df_cleaned.groupby(["Order Year","Order Month"])["Sales"]
    .sum()
    .sort_index()
)
print("Monthly Revenue Trend:")
print(monthly_revenue)
```
